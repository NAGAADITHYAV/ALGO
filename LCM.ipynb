```python
import random
```

```python
def GCD(a,b):
    if b==0:
        return a
    else :
        return GCD(b,a%b)
```

```python
def lcm_naive(a, b):
    for l in range(1, a*b + 1):
        if l % a == 0 and l % b == 0:
            return l

    return a*b
```

```python
a,b=map(int,input().split())
if b>a:
    temp=a
    a=b
    b=temp

gcd=GCD(a,b)    
lcm=a*b/gcd
print(lcm)
lcm=lcm_naive(a,b)
print(lcm)
```

```python
while(True):
    a=random.randint(0,200)
    b=random.randint(0,200)
    lcm=int(a*b/GCD(a,b))
    lcmn=lcm_naive(a,b)
    if lcm==lcmn :
        print("OK")
    else :
        print(a,b)
        print(lcm,lcmn)
        break
```
